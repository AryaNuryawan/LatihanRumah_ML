```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
df = pd.read_excel(path +'/Data/hasil_prediksi.xlsx')
df.head()
```

```python
y_true = df['Hasil Sebenarnya'].values
y_pred = df['Hasil Prediksi'].values
```

```python
cm = confusion_matrix(y_true, y_pred, labels=['Lulus', 'Tidak Lulus'])
print('Confusion Matrix:')
print(cm)
```

```python
acc = accuracy_score(y_true, y_pred)
print(f'Accuracy: {acc * 100:.2f}%')
```

```python
precision = precision_score(y_true, y_pred, pos_label='Lulus')
recall = recall_score(y_true, y_pred, pos_label='Lulus')
print(f'Precision (Lulus): {precision * 100:.2f}%')
print(f'Recall (Lulus): {recall * 100:.2f}%')
```
